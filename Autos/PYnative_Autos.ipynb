```python
# Import dependencies
import pandas as pd
import numpy as np

# Load CSV
auto_data_to_load = "Resources/Automobile_data.csv"

# Read CSV
auto_data_df = pd.read_csv(auto_data_to_load)
```

```python
# Print first five rows
auto_data_df.head()
```

```python
# Print last five rows
auto_data_df.tail()
```

```python
# Replace all column values that contain '?' and 'n.a.' with NaN
auto_data_df.loc[(auto_data_df["price"] == "")] = np.nan
auto_data_df
```

```python
# Show only the most expensive car and its price
most_expensive = auto_data_df[['company', 'price']][auto_data_df.price == auto_data_df['price'].max()]
most_expensive
```

```python
# Print all Toyota car details
toyota_details = auto_data_df.loc[auto_data_df['company'] == 'toyota']
toyota_details
```

```python
# Count total cars per company
car_count = auto_data_df['company'].value_counts()
car_count
```

```python
# Find each company's highest price car
car_mfg = auto_data_df.groupby('company')

company_highest_df = car_mfg['company','price'].max()
company_highest_df
```

```python
# Average miles of each car manufacturer
average_miles = car_mfg['company', 'average-mileage'].mean()
average_miles
```

```python
# Sort all cars by price 
top_5_most_expensive = auto_data_df.sort_values(["price"], ascending = False)
top_5_most_expensive.head()
```

```python
# Create two DFs using the following two dictionaries.  Concatonate those two DFs and create a key for each DF
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
german_cars_df = pd.DataFrame.from_dict(GermanCars)

japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}
japanese_cars_df = pd.DataFrame.from_dict(japaneseCars)

foreign_cars_df = pd.concat([german_cars_df, japanese_cars_df], keys=['Germany', 'Japan'])
foreign_cars_df
```

```python
# Create two DFs using the provided dictionaries.  Merge the two DFs and then append the second DF as a new colimn to the
# first DF
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
car_price_df = pd.DataFrame.from_dict(Car_Price)

car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
car_Horsepower_df = pd.DataFrame.from_dict(car_Horsepower)

new_auto = pd.merge(car_price_df, car_Horsepower_df, on=['Company'])
new_auto
```
