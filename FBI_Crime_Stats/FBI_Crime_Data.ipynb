```python
# Create new dataframe with Year and Population
#new = pd.DataFrame([old.A, old.B, old.C]).transpose()
pop_year_df = pd.DataFrame([crime_df.Year, crime_df.Population]).transpose()
pop_year_df
```

```python
# Set year as the index
# This is needed to have the year labels display across the x-axis
# If this isn't done, then the current index, 0-19, will display across the x-axis
pop_year_df.set_index('Year', inplace=True)
```

```python
# Create a bar chart 
x_labels=["Year"]
ax = pop_year_df.plot(kind='line', title='U.S. Population by Year', figsize=(15,10), 
                      legend=False, fontsize=12)

# Create labels for x- and y-axes
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Population', fontsize=12)

# Set the y limit between 0 and 45
ax.set_ylim(0, 350000000)
```

```python
# Create a bar chart 
x_labels=["Year"]
ax = pop_year_df.plot(kind='bar', title='U.S. Population by Year', figsize=(15,10), 
                      legend=False, fontsize=12)

# Create labels for x- and y-axes
ax.set_xlabel('Year', fontsize=12, rotation=45)
ax.set_ylabel('Population', fontsize=12)

# Set the y limit between 0 and 45
ax.set_ylim(0, 350000000)
```
