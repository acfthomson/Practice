```python
# Create new dataframe with Year and Population1
#new = pd.DataFrame([old.A, old.B, old.C]).transpose()
pop_year_df = pd.DataFrame([crime_df.Year, crime_df.Population1]).transpose()
pop_year_df
```

```python
# Create a bar chart using the MATLAB approach
ax = pop_year_df.plot(kind='bar', title='U.S. Population by Year', figsize=(15,10), legend=True, fontsize=12)
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Population', fontsize=12)
```
